```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

stock = yf.download('2330.tw', start = '2019-07-01', end = '2022-09-30')
stock.index.name = 'Date'
stock['Daily Return'] = stock['Adj Close'].pct_change().dropna()
stock = stock.drop(stock.index[0])
stock
```

```python
twii = yf.download('^TWII', start = '2019-07-01', end = '2022-09-30')
twii.index.name = 'Date'
twii['Daily Return'] = twii['Adj Close'].pct_change().dropna()
twii = twii.drop(twii.index[0])
twii
```

```python
rf = pd.read_csv('/Users/lofang/Desktop/python/csv files/Taiwan 10-Year Bond Yield 1907_2209.csv')
rf['Date'] = pd.to_datetime(rf['Date'])
rf
```

```python
rf = rf[rf["Date"].isin(twii.index)]
rf.set_index('Date', inplace = True)
rf
```

```python
fx_rate = pd.read_csv('/Users/lofang/Desktop/python/csv files/USD_TWD 1907_to_2209.csv')
fx_rate['Date'] = pd.to_datetime(fx_rate['Date'])
fx_rate
```

```python
fx_rate = fx_rate[fx_rate["Date"].isin(rf.index)]
fx_rate.set_index('Date', inplace=True)
fx_rate
```

```python
stock = stock[stock.index.isin(rf.index)]
twii = twii[twii.index.isin(rf.index)]
print(stock.info())
print(twii.info())
print(rf.info())
print(fx_rate.info())
```

```python
variable_table = pd.DataFrame(index = stock.index)
variable_table['Individual Return'] = stock['Daily Return']
variable_table['Weighted Return'] = twii['Daily Return']
variable_table['FX Factor'] = fx_rate['Price']
variable_table['risk-free'] = rf['Price']
variable_table['mkt-rf'] = variable_table['Weighted Return'] - variable_table['risk-free']
variable_table['risk premium'] = variable_table['Individual Return'] - variable_table['risk-free']
variable_table
```

```python
X = variable_table[['mkt-rf', 'FX Factor']]
y = variable_table['risk premium']
X = sm.add_constant(X)
fx_model = sm.OLS(y, X).fit()
print(fx_model.summary())
intercept, b1, b2 = fx_model.params
```
